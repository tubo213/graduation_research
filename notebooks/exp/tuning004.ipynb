```python
import yaml
from pathlib import Path
import copy
import os
from tqdm import tqdm
```

```python
YAML_DIR = Path('./yaml')
START_EXP_NUM = 206
```

```python
with open(YAML_DIR / 'optimize114.yaml', 'rb') as f:
    base_yaml = yaml.safe_load(f)
```

```python
lambda_list = [0, 0.0005, 0.001, 0.002, 0.003, 0.004, 0.005]
cluster_list = [5, 10, 50, 100]
# gamma_list = [0.2, 0.4, 0.6, 0.8]
```

```python
exp_no = START_EXP_NUM
for lam in tqdm(lambda_list, desc='lam'):
    for cluster in tqdm(cluster_list, desc='cluster', leave=False):
            exp_name = f'optimize{exp_no:03}'
            exp_yaml = copy.copy(base_yaml)
            name = f's-learner, previous_research, lambda={lam}, cluster={cluster}'
            exp_yaml['name'] = name
            exp_yaml['postprocess_config']['params']['n_cluster'] = cluster
            exp_yaml['postprocess_config']['params']['lambda'] = lam

            print(exp_yaml['name'])
            with open(YAML_DIR / f'{exp_name}.yaml', 'w') as f:
                yaml.dump(exp_yaml, f)
            
            command = f'python src/bin/optimize.py --exp {exp_name}'
            os.system(command)
            exp_no += 1
```
